```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
df = pd.read_csv("gold.csv")
```

```python
df
```

```python
df.describe()
```

```python
mean = df["Price"].mean()
mean
```

```python
df.plot(kind="scatter", x = 'Date', y = 'Price', s=0.5)
plt.axhline(y=mean, color="red", label="Mean") 
plt.legend(bbox_to_anchor=(0, 1), loc="upper left")
```

```python
arr = []
for item in df["Price"].values:
    arr.append(round((item-mean)**2, 2))

arr = np.array(arr)
df["Squared Deviations"]=arr
```

```python
arr = []
for item in df["Squared Deviations"].values:
    arr.append(round(math.sqrt(item), 2))

arr = np.array(arr)
df["Variance"]=arr
```

```python
df.plot(x="Date", y="Price", color="blue")
df.plot(x="Date", y="Squared Deviations", color="red")
df.plot(x="Date", y="Variance", color="green")
```
